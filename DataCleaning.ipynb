```python
product_data['BUY_UNIT'] = product_data['BUY_UNIT'].astype('category')
product_data['DESCRIPTION_STATUS'] = product_data['DESCRIPTION_STATUS'].astype('category')
product_data['BRAND'] = product_data['BRAND'].astype('category')
```

```python
product_data.select_dtypes(include = 'number').head()
```

```python
#Analyze null cases
product_data.isnull().sum()/len(product_data)*100
```

```python
Productdata2 = product_data[product_data['ORGANIC_ITEM'].isnull()]
Productdata2
```

```python
# Define columns to fill
columns_to_fill = ['BUY_UNIT', 'DESCRIPTION_STATUS', 'ORGANIC_ITEM']

for col in columns_to_fill:
    if col in product_data.columns and pd.api.types.is_categorical_dtype(product_data[col]):
        
        # Add 'N/A' to the category list
        product_data[col] = product_data[col].cat.add_categories(['N/A'])
    
    # Now fill NaN values with 'N/A'
    product_data[col] = product_data[col].fillna('N/A')
```

```python
product_data.isnull().sum()
```

```python
product_data['DESCRIPTION'] = product_data['DESCRIPTION'].replace({'<p>':'','</p>':''},regex = True)
product_data
```

```python
if 'DESCRIPTION' in product_data.columns:
    product_data['DESCRIPTION'] = product_data['DESCRIPTION'].fillna('')

    # Apply np.select
    product_data['Package'] = np.select(
        [
            product_data['DESCRIPTION'].str.contains("UN", case=False, na=False),
            product_data['DESCRIPTION'].str.contains("KG", case=False, na=False),
            product_data['DESCRIPTION'].str.contains("LT", case=False, na=False),
            product_data['DESCRIPTION'].str.contains("PZA", case=False, na=False),
            product_data['DESCRIPTION'].str.contains("ML", case=False, na=False),
            product_data['DESCRIPTION'].str.contains("GRS", case=False, na=False),
        ],
        ['UN', 'KG', 'LT', 'PZA', 'ML', 'GRS'],
        default='N/A'
    )
```

```python
del product_data['SUB_CATEGORY']
del product_data['SUB_SUB_CATEGORY']
```
